### For Loop

~~~python
dolphin_species = ["bottlenose dolphin","common dolphin","Amazon river dolphin",
                   "Ganges river dolphin","Commerson's dolphin","rough-toothed dolphin",
                   "Risso's dolphin"]

for i in range(len(dolphin_species)):
   print(dolphin_species[i], end = ' ') # all dolphin species printed
print()

for i in range(start = 0, stop = len(dolphin_species), step = 2):
   print(dolphin_species[i], end = ' ') # dolphin species at index 0 2 4 6 printed
print()

for species in dolphin_species:
   print(species) # print all dolphin species

for i, species in enumerate(dolphin_species):
   print(i, species) # print index followed by a species
~~~

### While Loop

You should use while loop when you're not sure how many times to iterate. For ex, if your character is moving in the world forever until a certain condition comes up to stop her.


### Display Dolphin via Predefined Classes & Libraries

We will read in the training csv file to a pandas dataframe, so we can get metadata we can display with our dolphin images from the training set.


```python
import pandas as pd

dolphin_df = pd.read_csv("train.csv", header='infer')
```

We can use libraries that have classes to list filenames from a directory, join a filename after directory name for the file we want to work with using **os** and even read in that image file as a image object using **mpimg**.

```python
import os
import matplotlib.image as mpimg

image_basepath = "train_images/"
image_filenames = os.listdir(image_basepath)
# image_filenames[0] = 000b8d89c738bd.jpg from train_images/ dir
image_filepath = os.path.join(image_basepath, image_filenames[0])
dolphin_image = mpimg.imread(image_filepath)
```

Once we have our image object, we can display it to the screen with extra info as needed using **plt**:

```python
import matplotlib.pyplot as plt
%matplotlib inline

# get image name from row 0, image column, check if it equals our filename
if dolphin_df.iloc[0]["image"] == image_filenames[0]:
    # Title of the image is image filepath and display image below title
    plt.suptitle(dolphin_df.iloc[0]["image"] + ": " + dolphin_df.iloc[0]["species"])
    plt.imshow(dolphin_image)
else:
    plt.suptitle(image_filepath)
    plt.imshow(dolphin_image)
```
